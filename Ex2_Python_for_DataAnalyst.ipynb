```python
%matplotlib inline
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv("datasets/tmdb_5000_movies.csv",sep=",")
```

```python
movies
```

```python
movies.count()
```

```python
#Affiche le temps cumulé d'heures de film pour chaque langues
movies.groupby("original_language").runtime.sum().sort_values()
```

```python
#Affiche les films dont les budgets dépassent ou sont égaux à 5M et les tries par ordre croissant
movies[movies.budget >=50000000].sort_values(by='budget')
```

```python
#Affiche les moyennes de vote et le nombre de film qui ont eux cette note par ordre croissant
movies.groupby("vote_average").vote_average.count().sort_values()
```

```python
#Evolution de la somme des budgets totale en fonction de l'année pour les films francais

plt.rcParams['figure.figsize'] = [15, 6] 
plt.rcParams['figure.dpi'] = 200 
movies[(movies.original_language=="fr")].groupby("release_date").budget.sum().plot()
```

```python
#Affiche les budgets en fonction de la date de sortie pour tous les films avec un budget supérieur à 100M 

movies[(movies.budget>=100000000)].groupby("release_date").budget.sum().plot()
```

```python
#Affiche le nombre de film en fonction de chaque notes à partir des films ayant au moins 3/10
movies[(movies.vote_average>=3)].groupby("vote_average").vote_average.count().plot()
```
